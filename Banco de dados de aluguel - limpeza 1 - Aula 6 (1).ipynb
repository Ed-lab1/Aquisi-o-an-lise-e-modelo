```python
# Queremos retirar uma string dos valores da coluna 'area', no caso o m2,  pra isso é necessário colocar entre pareteses e entre aspas
# olhar o site regexr.com
#df['area'] =
df['area'].str.extract('(.*) m²')
```

```python
df ['area'] = df['area'].str.extract('(.*) m²')
```

```python
df.head(2)
```

```python
# O ''(.*)'' pega tudo, exceto o que esta especificado fora do parenteses, neste caso a string vaga, que acompanha todos os valores da coluna
df ['vaga'] = df['vaga'].str.extract('(.*) Vaga')
```

```python
df ['banheiro'] = df['banheiro'].str.extract('(.*) Banheiros')
```

```python
df ['quartos'] = df['quartos'].str.extract('(.*) Quartos')
```

```python
df.head(3)
```

```python
# Vemos que o regex pegou tudo "(.*)" menos o que esta fora do parentes, antes e depois, No caso do R Sifrão ($), é necessário uma barra invertida antes do sifrão
#por ser um caractere especial
df ['aluguel'] = df['aluguel'].str.extract('R\$ (.*) /Mês ')
```

```python
df.head(2)
```

```python
# Na função replace os parametros são 1 -  O que se quer substituir, no caso o ponto '.'
# 2 - Pelo que se quer substituir, no caso nada, aspas vazias ('')
# 3 - regex = True , substituição de uma expressão regular pro outra expressão regula
df['aluguel'] = df['aluguel'].replace('\.' , '', regex=True)
```
